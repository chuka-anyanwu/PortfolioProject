The registration_year column seems to have some odd values as shown by the min and max the min value is 1000 way before the invention of cars and the max value is 9999 way into the future. 

Because a car can't be first registered after the listing was seen, any vehicle with a registration year above 2016 is definitely inaccurate. Determining the earliest valid year is more difficult. Realistically, it could be somewhere in the first few decades of the 1900s.

Let's count the number of listings with cars that fall outside the 1900 - 2016 interval and see if it's safe to remove those rows entirely, or if we need more custom logic.

We want to perform some aggregation operations on the brand column 

```python
brands = autos['brand'].unique()
```

```python
brands
```

```python
brands = autos['brand'].value_counts(normalize=True)
```

```python
brands
```

German cars seem to be the more popular cars (top 5) with Volkswagen being the most popular, we'll limit our analysis to brands with more than 5% of the total sales data.

```python
common_brands = brands[brands > 0.05].index
```

```python
common_brands
```

```python
# Create an empty dictionary
popular_brands = {}

for brands in common_brands:
    selected_brands = autos[autos['brand'] == brands]
    mean = selected_brands['price'].mean()
    popular_brands[brands] = int(mean)
```

```python
popular_brands
```

Based on the top 5 brands, Mercedes Benz seems to be the most expensive brand followed by Audi, then BMW, Opel is the least expensive and Volkswagen seems to be in-between hence could be the reason for it's popularity.

```python
mean_mileage = {}

for brands in common_brands:
    selected_brands = autos[autos['brand'] == brands]
    mileage = selected_brands['odometer_km'].mean()
    mean_mileage[brands] = int(mileage)
```

```python
mean_mileage
```

```python
pb_series = pd.Series(popular_brands).sort_values(ascending=False)
print(pb_series)
```

```python
df_mean = pd.DataFrame(pb_series, columns=['mean'])
df_mean
```

```python
mm_series = pd.Series(mean_mileage).sort_values(ascending=False)
print(mm_series)
```

```python
df_mileage = pd.DataFrame(mm_series, columns=['mileage'])
df_mileage
```

```python
df_mileage['mean'] = df_mean
df_mileage
```

Comparing both DataFrames it can be observed that the mileage seems to be related to the price of each brand, as the brands with the highest mileage are also the pricier brands i.e. Mercedes Benz, BMW and Audi. 

With Opel seeming to be the only outlier.
